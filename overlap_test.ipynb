```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
ti = 2

Lx = 5
Ly = 5

positions = np.loadtxt("data/positions_" + str(ti) + ".dat")
print(positions)
fig, ax = plt.subplots()

for p in positions:
    x = p[0]
    y = p[1]
    x -= np.floor(x / Lx) * Lx
    y -= np.floor(y / Ly) * Ly
    print(x, y)
    circle = plt.Circle((x, y), 1.0) 
    circle.set_alpha(0.3)
    ax.add_artist(circle)
    
ax.set_xlim([0,Lx])
ax.set_ylim([0,Ly])
ax.set_aspect(1)
```
