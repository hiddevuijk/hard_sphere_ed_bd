```python

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from sys import exit
```

```python
ti_max = 200
Lx = 10
Ly = Lx

colors = ["red", "blue", "green", "orange", "magenta", "brown", "darkblue"]
fig, ax = plt.subplots()
ax.set_xlim([0,Lx])
ax.set_ylim([0,Ly])
ax.set_aspect(1)

def init():
    return update_frame(0)

def update_frame(ti):
    if ti % 100 == 0: print(ti)
    ax.clear()
    ax.set_xlim([0,Lx])
    ax.set_ylim([0,Ly])
    ax.set_aspect(1)
    positions = np.loadtxt("data/positions_" + str(ti) + ".dat")
    for pi, p in enumerate(positions):
        col = colors[ pi % len(colors)]
        x = p[0]
        y = p[1]
        x -= np.floor(x / Lx) * Lx
        y -= np.floor(y / Ly) * Ly
        for dx in [-1,0,1]:
            for dy in [-1,0,1]:
                circle = plt.Circle((x + dx*Lx, y + dy*Ly), 0.5, color=col) 
                circle.set_alpha(0.3)
                ax.add_artist(circle)

anim = animation.FuncAnimation(fig, update_frame, init_func=init, frames=ti_max, interval=10)
anim.save("movie.mp4")

```
